**P(gender, Partner, PhoneService, OnlineSecurity, OnlineBackup | Churn)**

step1 : calculate prior >>
$P(Churn)$

```python
prior = df.groupby('Churn').size().div(len(df))
print(prior)
```

step2 : Calculating likelihood 

$P(gender|Churn)$

$P(Partner|Churn)$

$P(PhoneService|Churn)$

$P(OnlineSecurity|Churn)$

$P(OnlineBackup|Churn)$

```python
likelihood = {}
likelihood['gender'] = df.groupby(['Churn','gender']).size().div(len(df)).div(prior)
likelihood['Partner']= df.groupby(['Churn','Partner']).size().div(len(df)).div(prior)
likelihood['PhoneService'] = df.groupby(['Churn','PhoneService']).size().div(len(df)).div(prior)
likelihood['OnlineSecurity'] = df.groupby(['Churn','OnlineSecurity']).size().div(len(df)).div(prior)
likelihood['OnlineBackup'] = df.groupby(['Churn','OnlineBackup']).size().div(len(df)).div(prior)
print(likelihood)
```

**Calculating posterior**

{"gender":'male' , "Partner":'yes' , "PhoneService":'yes' , "OnlineSecurity":'no' , "OnlineBackup":'yes'}

```python
P_Yes = likelihood['gender']['Yes']['Male'] * likelihood['Partner']['Yes']['Yes'] * likelihood['PhoneService']['Yes']['Yes'] * likelihood['OnlineSecurity']['Yes']['No'] * likelihood['OnlineBackup']['Yes']['Yes'] * prior['Yes']
P_No = likelihood['gender']['No']['Male'] * likelihood['Partner']['No']['Yes'] * likelihood['PhoneService']['No']['Yes'] * likelihood['OnlineSecurity']['No']['No'] * likelihood['OnlineBackup']['No']['Yes'] * prior['No']


print(P_Yes)
print(P_No)
```
